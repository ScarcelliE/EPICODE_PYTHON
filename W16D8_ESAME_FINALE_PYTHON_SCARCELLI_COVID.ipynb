* Si chiede poi per ogni continente di trovare il numero di casi totali avvenuti in quello stesso continente; si chiede di
non considerare eventuali locazioni che nel dataset non appartengono ad alcun continente;

```python
# estraggo il massimo dei casi totali per localitÃ  e ne faccio la somma per continente, in quanto la colonna total_cases Ã¨ un progressivo
dfsng=dfsn.groupby(['continent','location']) 
dfmax=dfsng.total_cases.max()
dfmax.groupby('continent').sum()
```

```python
# faccio la somma dei nuovi casi per continente
dfg=dfsn.groupby("continent")
dfg["new_cases"].sum()
```

```python
dfg["new_cases"].sum().plot(kind="barh",grid=True)
```

Mi accorgo che per l'Europa i dati tra i due metodi coincidono 25.237.6812 di contagi, mentre per gli altri si differenziano di alcune migliaia di casi, mentre per l'Asia di quasi 32.000.
Esploro l'Asia e calcolo i progressivi dei nuovi casi e mi accorgo che il giorno 20-08-2023 per le Philippine c'Ã¨ un errore di 65.079 nei casi totali riportati, in particolare si registrano 4.108.552 casi contro 4.173.631 del progressivo effetivo di casi. Quindi per l'analisi considero la colonna dei nuovi casi facendone la somma.

```python
filtro_asia=dfsn.continent == "Asia"
dfsn.date=pd.to_datetime(df.date)
dfsn_asia=dfsn.loc[filtro_asia]
dfsn_asia["prog_new_cases"]=dfsn.groupby("location").new_cases.cumsum()
dfsn_asia=dfsn_asia.set_index("date")
dfsn_phi=dfsn_asia.loc[dfsn_asia.location=="Philippines"]
dfsn_phi.loc["2023-08-12":"2023-08-22",["location","total_cases","new_cases","prog_new_cases"]]
```

* Riguardo le nazioni di Italia, Germania e Francia, mostrare in un boxplot la differenza tra queste nazioni riguardo il
numero di pazienti in terapia intensiva (Intensive Care Unit, ICU) da maggio 2022 (incluso) ad aprile 2023 (incluso), e
scrivere un breve commento a riguardo;

```python
filtro_ita=dfsn.location=="Italy"
filtro_ger=dfsn.location=="Germany"
filtro_fra=dfsn.location=="France"
dfsn3n=dfsn.loc[(filtro_ita)|(filtro_ger)|(filtro_fra)]
filtro_maggio22=(dfsn3n.date.dt.year==2022)&(dfsn3n.date.dt.month>=5)
filtro_aprile23=(dfsn3n.date.dt.year==2023)&(dfsn3n.date.dt.month<=4)
dfsn3nd=dfsn3n.loc[(filtro_maggio22)|(filtro_aprile23)]
sns.boxplot(data=dfsn3nd,x="location",y="icu_patients")
```

*Distribuzione di ricoveri giornialieri in terapia intensiva (ICU) da maggio 2022 ad aprile 2023 per Francia-Germania-Italia*

Si evince che nel periodo indicato in Italia la situazione di ricoveri in terapia intensiva Ã¨ stata migliore rispetto alla Francia e alla Germania. In Germania si Ã¨ raggiunto un picco massimo di 1842 pazienti in ICU, anche se in generale con la Francia si evincono valori pressochÃ¨ uguali di mediana (994 contro i 972 pazienti) e di media (1022,39 contro i 998,38). Inoltre da notare che la media Ã¨ vicina alla mediana quindi abbiamo distribuzione quasi simmmetrica di tipo gaussiano.
